### Profesionales de la Salud y Cuidado
## Dentista: Número de colegiado, clínica propia, servicios ofrecidos, materiales en stock, citas del día y años de práctica.

```python
# Inicializamos el diccionario vacío
Diccionario_dentista = {}

# --- FUNCIÓN PARA AGREGAR DATOS INICIALES (4 Registros) ---
def carga_inicial():
    Diccionario_dentista["Rafael"] = {
        "Colegiado": "D-203", "Clinica": "Si", "Servicios": "Limpieza Dental",
        "Materiales": "Resina", "Citas": 8, "Anios": 5
    }
    Diccionario_dentista["Dylan"] = {
        "Colegiado": "D-145", "Clinica": "Si", "Servicios": "Ortodoncia",
        "Materiales": "Brackets", "Citas": 12, "Anios": 7
    }
    Diccionario_dentista["Maria"] = {
        "Colegiado": "D-310", "Clinica": "No", "Servicios": "Endodoncia",
        "Materiales": "Anestesia", "Citas": 6, "Anios": 4
    }
    Diccionario_dentista["Carlos"] = {
        "Colegiado": "D-411", "Clinica": "Si", "Servicios": "Blanqueamiento",
        "Materiales": "Gel Dental", "Citas": 10, "Anios": 9
    }

# --- FUNCIONES DEL SISTEMA ---

def agregar_dentista():
    nombre = input("\nNombre del Dentista: ").capitalize()
    datos = {
        "Colegiado": input("Número de colegiado: "),
        "Clinica": input("Clínica propia (Si/No): "),
        "Servicios": input("Servicios ofrecidos: "),
        "Materiales": input("Materiales en stock: "),
        "Citas": input("Citas del día: "),
        "Anios": input("Años de práctica: ")
    }
    Diccionario_dentista[nombre] = datos
    print(f"✅ ¡{nombre} ha sido registrado en el sistema!")

def listar_dentistas():
    if not Diccionario_dentista:
        print("\nNo hay dentistas registrados.")
    else:
        print("\n" + "="*115)
        print(f"{'DENTISTA':<12} | {'COLEGIADO':<12} | {'CLINICA':<10} | {'SERVICIOS':<18} | {'MATERIALES':<15} | {'CITAS':<6} | {'AÑOS'}")
        print("-" * 115)
        # Uso de for clave, valor
        for nombre, info in Diccionario_dentista.items():
            print(f"{nombre:<12} | {info['Colegiado']:<12} | {info['Clinica']:<10} | {info['Servicios']:<18} | {info['Materiales']:<15} | {info['Citas']:<6} | {info['Anios']}")
        print("="*115)

def buscar_dentista():
    nombre = input("\nNombre del dentista a buscar: ").capitalize()
    if nombre in Diccionario_dentista:
        p = Diccionario_dentista[nombre]
        print(f"\n🔍 DATOS DE {nombre.upper()}:")
        for k, v in p.items():
            print(f"- {k}: {v}")
    else:
        print("❌ El dentista no existe.")

def borrar_dentista():
    nombre = input("\nNombre del dentista a eliminar: ").capitalize()
    if nombre in Diccionario_dentista:
        del Diccionario_dentista[nombre]
        print(f"🗑️ Registro de {nombre} eliminado.")
    else:
        print("❌ No se encontró el registro.")

def actualizar_dentista():
    nombre = input("\nNombre del dentista a actualizar: ").capitalize()
    if nombre in Diccionario_dentista:
        print("Deje en blanco para mantener el dato actual.")
        for k in Diccionario_dentista[nombre]:
            nuevo_valor = input(f"Nuevo valor para {k} (Actual: {Diccionario_dentista[nombre][k]}): ")
            if nuevo_valor:
                Diccionario_dentista[nombre][k] = nuevo_valor
        print("📝 Registro actualizado con éxito.")
    else:
        print("❌ El dentista no existe.")

# --- EJECUCIÓN ---

carga_inicial()

while True:
    print("\n+---------------------------------------+")
    print("|      GESTIÓN DE DENTISTAS EXPERTOS    |")
    print("+--------+------------------------------+")
    print("| OPCIÓN |         DESCRIPCIÓN          |")
    print("+--------+------------------------------+")
    print("|   1    | Agregar nuevo Dentista       |")
    print("|   2    | Listar todos los Dentistas   |")
    print("|   3    | Buscar por nombre            |")
    print("|   4    | Borrar registro              |")
    print("|   5    | Actualizar información       |")
    print("|   6    | Salir del programa           |")
    print("+--------+------------------------------+")

    opcion = input("Seleccione (1-6): ")

    if opcion == "1":
        agregar_dentista()
    elif opcion == "2":
        listar_dentistas()
    elif opcion == "3":
        buscar_dentista()
    elif opcion == "4":
        borrar_dentista()
    elif opcion == "5":
        actualizar_dentista()
    elif opcion == "6":
        print("\n¡Cerrando consultorios! Hasta pronto.")
        break
    else:
        print("\n⚠ Opción inválida.")
```
