### Profesionales de la Salud y Cuidado
## Dentista: Número de colegiado, clínica propia, servicios ofrecidos, materiales en stock, citas del día y años de práctica.

```python
# Diccionario vacío
diccionario_dentista = []

# Datos iniciales
diccionario_dentista.append({"Colegiado":"101","Clinica":"Dental Sonrisa","Servicios":"Limpieza","Materiales":"Resina","Citas":"5","Años":"8"})
diccionario_dentista.append({"Colegiado":"102","Clinica":"Smile Center","Servicios":"Ortodoncia","Materiales":"Brackets","Citas":"7","Años":"10"})
diccionario_dentista.append({"Colegiado":"103","Clinica":"DentiCare","Servicios":"Extracción","Materiales":"Anestesia","Citas":"4","Años":"6"})
diccionario_dentista.append({"Colegiado":"104","Clinica":"White Teeth","Servicios":"Blanqueamiento","Materiales":"Gel","Citas":"6","Años":"9"})


# Listar tabla perfecta
def listar_dentista():

    print("\n" + "-" * 82)
    print("{:<12}{:<18}{:<18}{:<15}{:<10}{:<8}".format(
        "Colegiado", "Clinica", "Servicios", "Materiales", "Citas", "Años"))
    print("-" * 82)

    for d in diccionario_dentista:
        print("{:<12}{:<18}{:<18}{:<15}{:<10}{:<8}".format(
            d["Colegiado"],
            d["Clinica"],
            d["Servicios"],
            d["Materiales"],
            d["Citas"],
            d["Años"]
        ))

    print("-" * 82)


# Agregar
def agregar_dentista():

    c = input("Colegiado: ")
    cl = input("Clínica: ")
    s = input("Servicios: ")
    m = input("Materiales: ")
    ci = input("Citas: ")
    a = input("Años: ")

    diccionario_dentista.append({
        "Colegiado": c,
        "Clinica": cl,
        "Servicios": s,
        "Materiales": m,
        "Citas": ci,
        "Años": a
    })

    listar_dentista()


# Buscar
def buscar_dentista():

    c = input("Buscar colegiado: ")

    for d in diccionario_dentista:
        if d["Colegiado"] == c:
            print("\nRegistro encontrado:")
            print("-" * 40)
            for key, value in d.items():
                print(key + ":", value)
            print("-" * 40)

    listar_dentista()


# Borrar
def borrar_dentista():

    c = input("Borrar colegiado: ")

    for d in diccionario_dentista:
        if d["Colegiado"] == c:
            diccionario_dentista.remove(d)
            print("Registro eliminado")

    listar_dentista()


# Actualizar
def actualizar_dentista():

    c = input("Actualizar colegiado: ")

    for d in diccionario_dentista:
        if d["Colegiado"] == c:
            d["Clinica"] = input("Nueva clínica: ")
            d["Servicios"] = input("Nuevo servicio: ")
            print("Registro actualizado")

    listar_dentista()


# Menú
while True:

    print("\n1 Agregar")
    print("2 Listar")
    print("3 Buscar")
    print("4 Borrar")
    print("5 Actualizar")
    print("6 Salir")

    op = input("Opción: ")

    if op == "1":
        agregar_dentista()

    elif op == "2":
        listar_dentista()

    elif op == "3":
        buscar_dentista()

    elif op == "4":
        borrar_dentista()

    elif op == "5":
        actualizar_dentista()

    elif op == "6":
        print("Programa finalizado")
        break

    else:
        print("Opción incorrecta")
```
